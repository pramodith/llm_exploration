## Exploring variance of query and key vectors and the effect of scaling dot products in self-attention mechanism

```python
import numpy as np
```

```python
hidden_dim = np.random.randint(256, 1024)
print(f"Hidden dim: {hidden_dim}")
query = np.random.randn(1, hidden_dim)
key = np.random.randn(8, hidden_dim)

print(f"Query var: {query.var()}, Key var: {key.var()}")
```

```python
dot_product = np.matmul(query, key.T)
print(f"Dot product var (before scaling): {dot_product.var()}")
```

Variance remains around 1 after scaling the dot products by the square root of the hidden dimension.

```python
scaled_dot_product = dot_product / np.sqrt(hidden_dim)
print(f"Dot product var (after scaling): {scaled_dot_product.var()}")
```
